```
import cv2
import numpy as np
import matplotlib.pyplot as plt

from skimage import filters
from skimage import feature
```

```
img = np.array(cv2.imread("bridge1.jpg", cv2.IMREAD_GRAYSCALE))
imgd = cv2.normalize(img.astype("float"), None, 0.0, 1.0, cv2.NORM_MINMAX)

# sobel
sobel = (filters.sobel(imgd))

# canny
canny = feature.canny(imgd)

# prewitt
prewitt = filters.prewitt(imgd)

# roberts
roberts = filters.roberts(imgd)
  

fig, axes= plt.subplots(2, 2, figsize=(10, 8))
axes[0,0].imshow(sobel, cmap='gray')
axes[0,0].set_title('sobel filter')

axes[0,1].imshow(canny, cmap='gray')
axes[0,1].set_title('canny filter')

axes[1,0].imshow(prewitt, cmap='gray')
axes[1,0].set_title('prewitt filter')

axes[1,1].imshow(roberts, cmap='gray')
axes[1,1].set_title('roberts filter')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

cv2.waitKey(0)
cv2.destroyAllWindows()
```
