```
import numpy as np
import matplotlib.pyplot as plt

from skimage import data
from skimage.color import rgb2hsv
```

```
rgb_img = data.coffee()

cmy_img = 255-rgb_img
c_channel = cmy_img[:, :, 0]
m_channel = cmy_img[:, :, 1]
y_channel = cmy_img[:, :, 2]


fig, axes= plt.subplots(2, 2, figsize=(12, 8))
axes[0,0].imshow(rgb_img)
axes[0,0].set_title('rgb image')

axes[0,1].imshow(c_channel, cmap='gray')
axes[0,1].set_title('c_channel')

axes[1,0].imshow(m_channel, cmap='gray')
axes[1,0].set_title('m_channel')

axes[1,1].imshow(y_channel, cmap='gray')
axes[1,1].set_title('y_channel')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])


plt.show()

```
