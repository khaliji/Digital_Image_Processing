```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
# Read the image in greyscale
image = cv2.imread("beach.jpg", cv2.IMREAD_GRAYSCALE)
# In python if we want to perform matrix operations, we should use numpy array
p = np.array(image, dtype=np.float64)

# Contrast stretching formula is:
# p = T(q) = 1 / (1 + (m / q)^S
S = 0.9
m = 60

# We need epsilon value to prevent 'divied by zero' error
eps = np.spacing(1.0)
q = 1.0/(1.0+(m/(p + eps))**S)


fig, axes= plt.subplots(1, 2, figsize=(20, 8))

axes[0].imshow(q, cmap='gray')
axes[0].set_title('orginal image')

axes[1].imshow(p, cmap='gray')
axes[1].set_title('improved image')

# remove the x and y ticks
for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
