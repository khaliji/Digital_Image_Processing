```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
# Read the image in greyscale
img = np.array(cv2.imread("beach.jpg", cv2.IMREAD_GRAYSCALE), dtype=np.float64)
norm_img = cv2.normalize(src=img, dst=None, alpha=0.0, beta=1.0, norm_type=cv2.NORM_MINMAX)
completed = 1-norm_img

fig, axes= plt.subplots(1, 2, figsize=(20, 8))

axes[0].imshow(norm_img, cmap='gray')
axes[0].set_title('orginal image')

axes[1].imshow(completed, cmap='gray')
axes[1].set_title('completed image')

# remove the x and y ticks
for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
