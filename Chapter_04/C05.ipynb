```
import cv2
import skimage
from skimage.filters.rank import median
from skimage.morphology import disk
import numpy as np
import matplotlib.pyplot as plt
```

```

img = cv2.imread("azadi_tower.jpg", cv2.IMREAD_GRAYSCALE)
# Apply salt & pepper noise
noisy_img = skimage.util.random_noise(img, mode='s&p')
# Foe remove below warning use below code instead of above code
# noisy_img = np.uint8(skimage.util.random_noise(img, mode='s&p')*255)

fig, axes= plt.subplots(2, 2, figsize=(10, 8))

axes[0,0].imshow(noisy_img, 'gray')
axes[0,0].set_title('noisy image')

axes[0,1].imshow(median(noisy_img, disk(3)), 'gray')
axes[0,1].set_title('Kernel size = 3*3')

axes[1,0].imshow(median(noisy_img, disk(5)), 'gray')
axes[1,0].set_title('Kernel size = 5*5')

axes[1,1].imshow(median(noisy_img, disk(7)), 'gray')
axes[1,1].set_title('Kernel size = 7*7')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])


plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
