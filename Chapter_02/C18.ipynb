```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = np.array(cv2.imread("bridge1.jpg", cv2.IMREAD_GRAYSCALE))
imgd = cv2.normalize(img.astype("float"), None, 0.0, 1.0, cv2.NORM_MINMAX)

# set kernel size
s1 = (3,3)
s2 = (7,7)

# set sigma value
v1 = 0.5
v2 = 1
v3 = 2
v4 = 7
  
# Use cv2.GaussianBlur() as Gaussian filter 
img1 = cv2.GaussianBlur(imgd, s1, v1, cv2.BORDER_REPLICATE)  
img2 = cv2.GaussianBlur(imgd, s1, v2, cv2.BORDER_REPLICATE) 
img3 = cv2.GaussianBlur(imgd, s2, v3, cv2.BORDER_REPLICATE) 
img4 = cv2.GaussianBlur(imgd, s2, v4, cv2.BORDER_REPLICATE) 

fig, axes= plt.subplots(2, 2, figsize=(10, 8))
axes[0,0].imshow(img1, cmap='gray')
axes[0,0].set_title('sigma={}  , kernel_size={}'.format(v1, s1))

axes[0,1].imshow(img2, cmap='gray')
axes[0,1].set_title('sigma={}  , kernel_size={}'.format(v2, s1))

axes[1,0].imshow(img3, cmap='gray')
axes[1,0].set_title('sigma={}  , kernel_size={}'.format(v3, s2))

axes[1,1].imshow(img4, cmap='gray')
axes[1,1].set_title('sigma={}  , kernel_size={}'.format(v4, s2))

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

cv2.waitKey(0)
cv2.destroyAllWindows()
```
