```
import numpy as np
import matplotlib.pyplot as plt

from skimage import data
from skimage.color import rgb2lab
```

```
rgb_img = data.coffee()
lab_img = rgb2lab(rgb_img)

l_channel = lab_img[:, :, 0]
a_channel = lab_img[:, :, 1]
b_channel = lab_img[:, :, 2]


fig, axes= plt.subplots(2, 2, figsize=(12, 8))
axes[0,0].imshow(rgb_img)
axes[0,0].set_title('rgb image')

axes[0,1].imshow(l_channel, cmap='gray')
axes[0,1].set_title('l_channel')

axes[1,0].imshow(a_channel, cmap='gray')
axes[1,0].set_title('a_channel')

axes[1,1].imshow(b_channel, cmap='gray')
axes[1,1].set_title('b_channel')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])


plt.show()

```
