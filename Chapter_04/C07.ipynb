```
import cv2
import numpy as np
import matplotlib.pyplot as plt

import skimage
from scipy import ndimage
```

```
img = cv2.imread("azadi_tower.jpg", cv2.IMREAD_GRAYSCALE)
# Apply salt & pepper noise
noisy_img = skimage.util.random_noise(img, mode='s&p')

imgd1 = ndimage.maximum_filter(img, size=3)
imgd2 = ndimage.minimum_filter(img, size=3)

fig, axes= plt.subplots(1, 3, figsize=(15, 8))

axes[0].imshow(noisy_img, 'gray')
axes[0].set_title('noisy image')

axes[1].imshow(imgd1, 'gray')
axes[1].set_title('maximum_filter 3*3')

axes[2].imshow(imgd2, 'gray')
axes[2].set_title('minimum_filter 3*3')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])


plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
