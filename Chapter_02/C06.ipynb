```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = np.zeros((256, 256))
img[104:149, 104:149] = 1

# Fourier transform
imgF = np.fft.fft2(img)
imgF_shift = np.fft.fftshift(imgF)
imgF_shift_d = np.abs(imgF_shift)
img_log = np.log(1+imgF_shift_d)
max_val = np.max(img_log)


fig, axes= plt.subplots(1, 2, figsize=(10, 10))
axes[0].imshow(img, 'gray')
axes[0].set_title('orginal image')

axes[1].imshow(img_log/max_val, cmap='gray', vmin=np.min(img_log/max_val), vmax=np.max(img_log/max_val))
axes[1].set_title('fftshift image')

# remove the x and y ticks
for i in range(axes.shape[0]):
    axes[i].set_xticks([])
    axes[i].set_yticks([])
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
