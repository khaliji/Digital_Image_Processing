```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = np.array(cv2.imread("jungle.png", cv2.IMREAD_GRAYSCALE))
imgd = cv2.normalize(img.astype("float"), None, 0.0, 1.0, cv2.NORM_MINMAX)

# Average filter
Iavr = cv2.blur(imgd, ksize=(3, 3), borderType=cv2.BORDER_REPLICATE)
g = imgd.astype('float')-Iavr.astype('float')
k = 2
Ip = imgd.astype('float') + k*g

fig, axes= plt.subplots(1, 2, figsize=(15, 5))
axes[0].imshow(imgd, cmap='gray')
axes[0].set_title('original image')

axes[1].imshow(Ip, cmap='gray')
axes[1].set_title('sharpped image')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

cv2.waitKey(0)
cv2.destroyAllWindows()
```
