```
import cv2
import skimage
import numpy as np
import matplotlib.pyplot as plt
```

```

img = np.array(cv2.imread("azadi_tower.jpg", cv2.IMREAD_GRAYSCALE))
# Apply speckle noise with var=0.04
noisy_img = skimage.util.random_noise(img, mode='speckle', var=0.04)

fig, axes= plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img, 'gray')
axes[0].set_title('original image')

axes[1].imshow(noisy_img, 'gray')
axes[1].set_title('noisy image')

for i in range(axes.shape[0]):
    axes[i].set_xticks([])
    axes[i].set_yticks([])


plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
