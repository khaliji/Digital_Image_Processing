```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = np.array(cv2.imread("bridge1.jpg", cv2.IMREAD_GRAYSCALE))
imgd = cv2.normalize(img.astype("float"), None, 0.0, 1.0, cv2.NORM_MINMAX)

# set kernel size
s1 = (3,3)
s2 = (7,7)
  
# Use cv2.medianBlur as median filter 
img1 = cv2.GaussianBlur(imgd, s1, cv2.BORDER_REPLICATE)  
img2 = cv2.GaussianBlur(imgd, s2, cv2.BORDER_REPLICATE) 

fig, axes= plt.subplots(1, 2, figsize=(10, 8))
axes[0].imshow(img1, cmap='gray')
axes[0].set_title('median filter with kernel_size={}'.format(s1))

axes[1].imshow(img2, cmap='gray')
axes[1].set_title('median filter with kernel_size={}'.format(s2))

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

cv2.waitKey(0)
cv2.destroyAllWindows()
```
