```
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
# Read the image in greyscale
img = cv2.imread(filename="Imam.jpg", flags=cv2.IMREAD_GRAYSCALE)

fig, axes = plt.subplots(3, 3, figsize=(18,9))
# mask is 3-dim matrix of 8 planes with original image size
# Each plane is array with values 2^i, i is index of plane
mask = np.ones((8, img.shape[0], img.shape[1]), dtype=np.uint8)
for i in range(8):
  mask[i] *= (2**i)

# Show planes
for i in range(8):
  plane = (cv2.bitwise_and(mask[i], img))  
  axes[i//3,i%3].imshow(plane, cmap='gray')
  axes[i//3,i%3].set_title("plane_{}".format(i+1))
  axes[i//3,i%3].set_xticks([])
  axes[i//3,i%3].set_yticks([])

#Show original page
axes[2,2].imshow(img, cmap='gray')
axes[2,2].set_title("original image")
axes[2,2].set_xticks([])
axes[2,2].set_yticks([])

plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
