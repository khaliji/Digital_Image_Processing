```
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```
img = cv2.imread(filename="coins.jpg", flags=cv2.IMREAD_GRAYSCALE)

# kernel
kernel = cv2.getStructuringElement(shape=cv2.MORPH_RECT, ksize=(2,2))

# Internal edge
imgO = cv2.erode(img, kernel ,iterations = 1)
EI = cv2.subtract(img,imgO)

# External edge
imgD = cv2.dilate(img, kernel ,iterations = 1)
EO = cv2.subtract(imgD, img)

# Gradient
EG = cv2.subtract(imgD, imgO)


fig, axes = plt.subplots(2,2, figsize=(12,10))
axes[0,0].imshow(img, cmap='gray')
axes[0,0].set_title("image")

axes[0,1].imshow(EI, 'gray')
axes[0,1].set_title("Internal edge")

axes[1,0].imshow(EO, 'gray')
axes[1,0].set_title("External edge")

axes[1,1].imshow(EG, 'gray')
axes[1,1].set_title("Gradient")

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```
