```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = np.array(cv2.imread("bridge1.jpg", cv2.IMREAD_GRAYSCALE))
imgd = cv2.normalize(img.astype("float"), None, 0.0, 1.0, cv2.NORM_MINMAX)

# set kernel
gx = np.array([[-1, -2, -1],[0, 0, 0],[1, 2, 1]])
gy = gx.T
  
# Apply sobel filter
Imggx = cv2.filter2D(imgd, -1, gx,borderType=cv2.BORDER_REPLICATE)
Imggy = cv2.filter2D(imgd, -1, gy,borderType=cv2.BORDER_REPLICATE)

M = np.sqrt(Imggx**2 + Imggy**2)

fig = plt.figure(figsize=(5,5))
plt.imshow(M.astype("uint8"), cmap='gray')
plt.title("Both horizontal and vertical sobel filtering")
plt.xticks([])
plt.yticks([])
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
