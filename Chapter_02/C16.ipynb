```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = np.array(cv2.imread("bridge1.jpg", cv2.IMREAD_GRAYSCALE))
imgd = cv2.normalize(img.astype("float"), None, 0.0, 1.0, cv2.NORM_MINMAX)

# set kernel size
h1 = (3,3)
h2 = (7,7)
h3 = (21,21)
  
# Using cv2.blur() as average filter method 
img1 = cv2.blur(imgd, h1, cv2.BORDER_REPLICATE) 
img2 = cv2.blur(imgd, h2, cv2.BORDER_REPLICATE)
img3 = cv2.blur(imgd, h3, cv2.BORDER_REPLICATE)

fig, axes= plt.subplots(2, 2, figsize=(10, 8))
axes[0,0].imshow(imgd, cmap='gray')
axes[0,0].set_title('original')

axes[0,1].imshow(img1, cmap='gray')
axes[0,1].set_title('mean filter(3*3)')

axes[1,0].imshow(img2, cmap='gray')
axes[1,0].set_title('mean filter(7*7)')

axes[1,1].imshow(img3, cmap='gray')
axes[1,1].set_title('mean filter(21*21)')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

cv2.waitKey(0)
cv2.destroyAllWindows()
```
