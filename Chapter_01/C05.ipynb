```
import cv2
import matplotlib.pyplot as plt
```

```
# Read the image in greyscale
image = cv2.imread('car.jpg',0)

# Resize image to half
img_resized = cv2.resize(image,None,None, fx=0.5, fy=0.5, interpolation=cv2.INTER_NEAREST)
# Restore image to original size
final_img_resized = cv2.resize(img_resized,None,None, fx=2, fy=2, interpolation=cv2.INTER_NEAREST)

# We set 'sharex' and 'sharey' to True, so we can show each image with actual size
fig, axes= plt.subplots(1, 3, figsize=(20, 8), sharex=True, sharey=True)

axes[0].imshow(image, 'gray')
axes[0].set_title('orginal')

axes[1].imshow(img_resized, 'gray')
axes[1].set_title('img_resized')

axes[2].imshow(final_img_resized, 'gray')
axes[2].set_title('final_img_resized')

for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
