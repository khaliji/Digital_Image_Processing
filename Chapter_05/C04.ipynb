```
import numpy as np
import matplotlib.pyplot as plt

from skimage import data
from skimage.color import rgb2yiq
```

```
rgb_img = data.coffee()
yiq_img = rgb2yiq(rgb_img)

y_channel = yiq_img[:, :, 0]
i_channel = yiq_img[:, :, 1]
q_channel = yiq_img[:, :, 2]


fig, axes= plt.subplots(2, 2, figsize=(12, 8))
axes[0,0].imshow(rgb_img)
axes[0,0].set_title('rgb image')

axes[0,1].imshow(y_channel, cmap='gray')
axes[0,1].set_title('y_channel')

axes[1,0].imshow(i_channel, cmap='gray')
axes[1,0].set_title('i_channel')

axes[1,1].imshow(q_channel, cmap='gray')
axes[1,1].set_title('q_channel')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])


plt.show()

```
