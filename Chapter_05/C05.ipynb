```
import numpy as np
import matplotlib.pyplot as plt

from skimage import data
from skimage.color import rgb2ycbcr
```

```
rgb_img = data.coffee()
ycbcr_img = rgb2ycbcr(rgb_img)

y_channel = ycbcr_img[:, :, 0]
cb_channel = ycbcr_img[:, :, 1]
cr_channel = ycbcr_img[:, :, 2]


fig, axes= plt.subplots(2, 2, figsize=(12, 8))
axes[0,0].imshow(rgb_img)
axes[0,0].set_title('rgb image')

axes[0,1].imshow(y_channel, cmap='gray')
axes[0,1].set_title('Y_channel')

axes[1,0].imshow(cb_channel, cmap='gray')
axes[1,0].set_title('Cb_channel')

axes[1,1].imshow(cr_channel, cmap='gray')
axes[1,1].set_title('Cr_channel')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])


plt.show()

```
