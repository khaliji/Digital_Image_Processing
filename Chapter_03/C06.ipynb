```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = np.zeros((256, 256))
img[104:149, 104:149] = 1

imgF = np.fft.fft2(img)
imgF_d = np.abs(imgF)
max_val_1 = np.max(imgF_d)

imgF_shift = np.fft.fftshift(imgF)
imgF_shift_d = np.abs(imgF_shift)
max_val_2 = np.max(imgF_shift_d)

fig, axes = plt.subplots(1, 3 ,figsize=(15,5))
axes[0].imshow(img, cmap='gray', vmin=np.min(img), vmax=np.max(img))
axes[0].set_title('original')

axes[1].imshow(imgF_d / max_val_1, cmap='gray', vmin=np.min(imgF_d / max_val_1), vmax=np.max(imgF_d / max_val_1))
axes[1].set_title('fft')

axes[2].imshow(imgF_shift_d/max_val_2, cmap='gray', vmin=np.min(imgF_shift_d/max_val_2), vmax=np.max(imgF_shift_d/max_val_2))
axes[2].set_title('fftshift')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()

fig, axes = plt.subplots(1, 2 ,figsize=(10,5))
axes[0].imshow(imgF_d / max_val_1, cmap='gray', vmin=np.min(imgF_d / max_val_1), vmax=np.max(imgF_d / max_val_1))
axes[0].set_title('fft')

axes[1].imshow(imgF_shift_d/max_val_2, cmap='gray', vmin=np.min(imgF_shift_d/max_val_2), vmax=np.max(imgF_shift_d/max_val_2))
axes[1].set_title('fftshift')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```
