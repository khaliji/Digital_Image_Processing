```
import cv2
import skimage
import numpy as np
import matplotlib.pyplot as plt
```

```
img = np.array(cv2.imread(filename="azadi_tower.jpg", flags=cv2.IMREAD_GRAYSCALE))
img_norm = np.array(cv2.normalize(img.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX))

[m, n] = img.shape
[a, b] = np.meshgrid(range(0, n), range(0, m))

periodic_noise = np.sin(a/2+b/4)+1
noisy_img = (img_norm + periodic_noise)/2

fig, axes= plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img, 'gray')
axes[0].set_title('original image')

axes[1].imshow(noisy_img, 'gray')
axes[1].set_title('noisy image')

for i in range(axes.shape[0]):
    axes[i].set_xticks([])
    axes[i].set_yticks([])


plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
