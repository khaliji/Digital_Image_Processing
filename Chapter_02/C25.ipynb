```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = np.array(cv2.imread("bridge1.jpg", cv2.IMREAD_GRAYSCALE))

# Create laplacian of a gaussian
# First apply gaussian filter
gaussian_img = cv2.GaussianBlur(img, ksize=(3, 3),sigmaX=2, borderType=cv2.BORDER_REPLICATE)
laplace_img = cv2.Laplacian(gaussian_img, ddepth=-1)

fig, axes= plt.subplots(1, 2, figsize=(15, 5))
axes[0].imshow(laplace_img, cmap='gray')
axes[0].set_title('LoG gray')

axes[1].imshow(laplace_img>laplace_img.max()/5, cmap='gray')
axes[1].set_title('LoG binary')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

cv2.waitKey(0)
cv2.destroyAllWindows()
```
