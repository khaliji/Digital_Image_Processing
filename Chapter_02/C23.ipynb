```
import cv2
import numpy as np
import matplotlib.pyplot as plt

from skimage import filters
from skimage import feature
```

```
img = np.array(cv2.imread("bridge1.jpg", cv2.IMREAD_GRAYSCALE))

# laplace
laplace_img =  cv2.Laplacian(img, ddepth=-1, borderType=cv2.BORDER_REPLICATE)

c = -1
sharpped_img = img + c*laplace_img

fig, axes= plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(img, cmap='gray')
axes[0].set_title('original')

axes[1].imshow(laplace_img.astype('uint8'), cmap='gray')
axes[1].set_title('laplace filter')

axes[2].imshow(sharpped_img.astype('uint8'), cmap='gray')
axes[2].set_title('sharpped image')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

cv2.waitKey(0)
cv2.destroyAllWindows()
```
