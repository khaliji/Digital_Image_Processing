```
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```
img = cv2.imread(filename="hand_write.png", flags=cv2.IMREAD_GRAYSCALE)
thresh_val, thresh_result = cv2.threshold(src=img, thresh=40, maxval=255, type=cv2.THRESH_BINARY)

binary_img = 255-thresh_result

# kernel
kernel = cv2.getStructuringElement(shape=cv2.MORPH_RECT, ksize=(3,3))
dilation = cv2.dilate(binary_img,kernel,iterations = 1)

fig, axes = plt.subplots(3, 1, figsize=(10,18))
# For displaying a grayscale image set up the color mapping using the parameters cmap='gray'.
axes[0].imshow(img, cmap='gray')
axes[0].set_title("Original")

axes[1].imshow(binary_img, 'gray')
axes[1].set_title("binary")

axes[2].imshow(dilation, 'gray')
axes[2].set_title("dilation")

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```
