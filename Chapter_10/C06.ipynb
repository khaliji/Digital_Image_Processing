```
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```
img = cv2.imread(filename="CT.jpg", flags=cv2.IMREAD_GRAYSCALE)

# kernel
kernel = cv2.getStructuringElement(shape=cv2.MORPH_ELLIPSE, ksize=(73,73))
imtop = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
imbot = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)

imfinal = cv2.subtract(cv2.add(img,imtop),imbot)

fig, axes = plt.subplots(1,4, figsize=(20,8))
axes[0].imshow(img, 'gray')
axes[0].set_title("original image")

axes[1].imshow(imtop, cmap='gray')
axes[1].set_title("TOP HAT")

axes[2].imshow(imbot, 'gray')
axes[2].set_title("BOTTOM HAT")

axes[3].imshow(imfinal, 'gray')
axes[3].set_title("Increased contrast")

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```
