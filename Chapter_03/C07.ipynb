```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = np.zeros((256, 256))
img[104:149, 104:149] = 1

imgF = np.fft.fft2(img)
imgF_shift = np.fft.fftshift(imgF)
imgF_shift_d = np.abs(imgF_shift)
img_log = np.log(1+imgF_shift_d)
max_val = np.max(img_log)

plt.imshow(img_log/max_val, cmap='gray', vmin=np.min(img_log/max_val), vmax=np.max(img_log/max_val))
plt.title('fftshift')
plt.xticks([]), plt.yticks([])

plt.show()
```
