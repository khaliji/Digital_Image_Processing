```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
# Read the image in greyscale
# We want to use matrix operations, so we use numpy array
img = np.array(cv2.imread("beach.jpg", cv2.IMREAD_GRAYSCALE), dtype=np.uint8)
img2 = np.copy(img)
img3 = np.copy(img)
completed = 255-img

# Complete only pixels that are more than 128.
img2[img2 > 128] = 255 - img2[img2 > 128]
# Complete only pixels that are less than 128.
img3[img3 < 128] = 255 - img3[img3 < 128]


fig, axes= plt.subplots(2, 2, figsize=(10, 10))
axes[0,0].imshow(img, 'gray')
axes[0,0].set_title('orginal')

axes[0,1].imshow(completed, 'gray')
axes[0,1].set_title('completed')

axes[1,0].imshow(img2, 'gray')
axes[1,0].set_title('more than 128')


axes[1,1].imshow(img3, 'gray')
axes[1,1].set_title('less than 128')

# remove the x and y ticks
for i in range(axes.shape[0]):
  for j in range(axes.shape[1]):
    axes[i,j].set_xticks([])
    axes[i,j].set_yticks([])
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
