```
import cv2
import numpy as np
import matplotlib.pyplot as plt

import skimage
from scipy.signal.signaltools import wiener
```

```
img = cv2.imread("azadi_tower.jpg", cv2.IMREAD_GRAYSCALE)
# Apply gaussian noise
noisy_img = skimage.util.random_noise(img, mode='gaussian', mean=0.0 ,var=0.001)
noisy_img = cv2.normalize(noisy_img.astype("float"), None, 0.0, 1.0, cv2.NORM_MINMAX)

imgd1 = wiener(noisy_img, (3, 3))
imgd2 = wiener(noisy_img, (5, 5))
imgd3 = wiener(noisy_img, (7, 7))

fig, axes= plt.subplots(2, 2, figsize=(10, 8))

axes[0,0].imshow(noisy_img, 'gray')
axes[0,0].set_title('noisy image')

axes[0,1].imshow(imgd1, 'gray')
axes[0,1].set_title('wiener(3*3) filtered')

axes[1,0].imshow(imgd2, 'gray')
axes[1,0].set_title('wiener(5*5) filtered')

axes[1,1].imshow(imgd3, 'gray')
axes[1,1].set_title('wiener(7*7) filtered')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])


plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
