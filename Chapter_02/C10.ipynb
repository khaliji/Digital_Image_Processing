```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = np.array(cv2.imread("professor.jpg", cv2.IMREAD_GRAYSCALE))

# Calc image histogram with 256 bins
Nhist = cv2.calcHist([img], [0], None, [256], [0, 256]) / np.size(img)

fig, axes= plt.subplots(1, 3, figsize=(15, 5))
axes[0].hist(img.ravel(), bins = np.arange(0,256))
axes[0].set_title('histogram with 256 bins')

axes[1].hist(img.ravel(), bins = np.arange(0,50))
axes[1].set_title('histogram with 50 bins')

axes[2].plot(np.arange(0,256), Nhist)
axes[2].set_title('normalized histogram')


cv2.waitKey(0)
cv2.destroyAllWindows()
```
