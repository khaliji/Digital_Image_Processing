```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = np.array(cv2.imread("bridge1.jpg", cv2.IMREAD_GRAYSCALE))
imgd = cv2.normalize(img.astype("float"), None, 0.0, 1.0, cv2.NORM_MINMAX)

# set kernel

gx = np.array([[-1, -2, -1],[0, 0, 0],[1, 2, 1]])
gy = gx.T
# sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)  # x
# sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)  # y
  
# Apply sobel filter
# cv2.filter2D (original image, -1 = the converted image has the same depth as the original image,
# convolution kernel, boundary type = REPLICATE)
Imggx = cv2.filter2D(imgd, -1, gx,borderType=cv2.BORDER_REPLICATE)
Imggy = cv2.filter2D(imgd, -1, gy,borderType=cv2.BORDER_REPLICATE)

fig, axes= plt.subplots(1, 2, figsize=(10, 8))
axes[0].imshow(Imggx.astype("uint8"), cmap='gray')
axes[0].set_title('Horizontal sobel filter')

axes[1].imshow(Imggy.astype("uint8"), cmap='gray')
axes[1].set_title('Vertical sobel filter')

for ax in axes.ravel():
    ax.set_xticks([])
    ax.set_yticks([])

cv2.waitKey(0)
cv2.destroyAllWindows()
```
