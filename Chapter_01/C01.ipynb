```
import cv2
import matplotlib.pyplot as plt
```

```
# cv2.IMREAD_GRAYSCALE: It specifies to load an image in grayscale mode.
# Alternatively, we can pass integer value 0 for this flag.
pic = cv2.imread(filename="Imam.jpg", flags=cv2.IMREAD_GRAYSCALE)

# cv2.threshold: This method returns two outputs.
# The first one is the threshold that was used and the second output is the thresholded image.
# If the pixel value is smaller than the threshold, it is set to 0, otherwise, it is set to a maximum value.
# cv2.THRESH_BINARY: If pixel intensity is greater than the set threshold, value set to 255, else set to 0 (black).
thresh_val, thresh_result = cv2.threshold(src=pic, thresh=127, maxval=255, type=cv2.THRESH_BINARY)

fig, axes = plt.subplots(1, 2, figsize=(10,4))
# For displaying a grayscale image set up the color mapping using the parameters cmap='gray'.
axes[0].imshow(pic, cmap='gray')
axes[0].set_title("Original")
axes[1].imshow(thresh_result, 'gray')
axes[1].set_title("binary")

# remove the x and y ticks
for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
