```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
def WCAE(img1, img2):
  [M,N] = img1.shape
  wcae = np.max(np.max(np.abs(img1-img2)))
  return wcae

img = np.array(cv2.imread(filename="azadi_tower.jpg", flags=cv2.IMREAD_GRAYSCALE))
img_norm = np.array(cv2.normalize(img.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX))


[m, n] = img.shape
[a, b] = np.meshgrid(range(0, n), range(0, m))

periodic_noise = np.sin(a+b/4)+1
noisy_img = (img_norm + periodic_noise)/2
noisy_img = np.array(cv2.normalize(noisy_img.astype('float'), None, 0.0, 255.0, cv2.NORM_MINMAX))

print("WCAE = {}".format(WCAE(img, noisy_img)))

fig, axes= plt.subplots(1, 2, figsize=(12, 8))
axes[0].imshow(img, cmap='gray')
axes[0].set_title('original image')

axes[1].imshow(noisy_img, cmap='gray')
axes[1].set_title('noisy image')

plt.show()

```
